# **Clustering-Hierarchical (Airlines)**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
from sklearn.preprocessing import normalize
```

```python
airline=pd.read_csv('EastWestAirlines..csv')
airline
```

```python
airline.info()
```

```python
airline2=airline.drop(['ID#'],axis=1)
airline2
```

```python
airline2_norm=pd.DataFrame(normalize(airline2),columns=airline2.columns)
airline2_norm
```

```python
plt.figure(figsize=(10, 7))  
dendograms=sch.dendrogram(sch.linkage(airline2_norm,'complete'))
```

```python
hclusters=AgglomerativeClustering(n_clusters=5,affinity='euclidean',linkage='ward')
hclusters
```

```python
y=pd.DataFrame(hclusters.fit_predict(airline2_norm),columns=['clustersid'])
y['clustersid'].value_counts()
```

```python
airline2['clustersid']=hclusters.labels_
airline2
```

```python
airline2.groupby('clustersid').agg(['mean']).reset_index()
```

```python
plt.figure(figsize=(10, 7))  
plt.scatter(airline2['clustersid'],airline2['Balance'], c=hclusters.labels_) 
```

# **K MEANS**

```python
from sklearn.cluster import KMeans
from scipy.spatial.distance import cdist
```

```python
Airline1=pd.read_csv('EastWestAirlines..csv')
```

```python
def norm_func(i):
    x=(i-i.min())/(i.max()-i.min())
    return(x)
```

```python
df_norm=norm_func(airline.iloc[:,1:])
```

```python
wcss=[]
for i in range(1,11):
    kmeans=KMeans(n_clusters=i)
    kmeans.fit(df_norm)
    wcss.append(kmeans.inertia_)
plt.plot(range(1,11),wcss)
plt.title('Elbow curv')
plt.xlabel('number of clusters')
plt.ylabel('WCSS')
plt.show()
```

```python
model=KMeans(n_clusters=4)
model.fit(df_norm)
model.labels_
```

```python
x=pd.Series(model.labels_)
Airline1['Clust']=x
```

```python
Airline1
```

```python
Airline1.iloc[:,1:5].groupby(Airline1.Clust).mean()
```

# **DBSCAN**

```python
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
```

```python
Airlines2=pd.read_csv('EastWestAirlines..csv')
```

```python
Airlines2.info()
```

```python
df=Airlines2.iloc[:,1:5]
df.values
```

```python
stscaler=StandardScaler().fit(df.values)
x=stscaler.transform(df.values)
```

```python
x
```

```python
dbscan=DBSCAN(eps=2,min_samples=5)
dbscan.fit(x)
```

```python
cl=pd.DataFrame(dbscan.labels_,columns=['cluster'])
cl
```

```python
pd.concat([Airlines2,cl],axis=1)
```
